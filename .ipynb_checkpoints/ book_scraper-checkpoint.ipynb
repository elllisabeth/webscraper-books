## Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import streamlit as st
```

## Create Scraper Function

```python
def scrape_books():
    base_url = "https://books.toscrape.com/catalogue/page-{}.html"
    book_list = []

    for page in range(1, 51):  # 50 pages total on website
        url = base_url.format(page)
        response = requests.get(url)

        if response.status_code != 200:
            break

        soup = BeautifulSoup(response.text, "html.parser")
        books = soup.find_all("article", class_="product_pod")

        for book in books:
            title = book.h3.a["title"]
            price = book.find("p", class_="price_color").text
            rating = book.p["class"][1]
            stock = book.find("p", class_="instock availability").text.strip()
            image_url = "https://books.toscrape.com/catalogue/" + book.img["src"].replace("../../", "")

            book_list.append({
                "title": title,
                "price": price,
                "rating": rating,
                "stock": stock,
                "image_url": image_url
            })

    return pd.DataFrame(book_list)
```

```python
if st.button("Run Scraper"):
    df = scrape_books()
    st.success("Scraping completed!")
    
    # Filter rating
    rating_filter = st.selectbox("Filter by rating:", ["All", "One", "Two", "Three", "Four", "Five"])
    if rating_filter != "All":
        df = df[df["rating"] == rating_filter]
    
    # Search bar
    search_query = st.text_input("Search book title:")
    if search_query:
        df = df[df["title"].str.contains(search_query, case=False)]
    
    st.dataframe(df)
```

## Create Scraper Function

```python
books = scrape_books()
books[:5]  # show sample
```

## Convert to DataFrame  

```python
df = pd.DataFrame(books)
df
```
